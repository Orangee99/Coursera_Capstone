<h2>Segmenting and Clustering Neighbourhoods in Toronto

Import Libraries Use pandas, or the BeautifulSoup package, or any other way you are comfortable with to transform the data in the table on the Wikipedia page into the above pandas dataframe.


```python
import pandas as pd
import numpy as np
import requests
```

```python
url='https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'

df=pd.read_html(url, header=0)[0]

df.head()
```

```python
df.drop(df[df['Borough']=='Not assigned'].index, inplace = True)
df.head()
```

```python
df
```

```python
## Neigbourhood are already grouped by Borough
df.groupby(['Postal Code','Borough']).agg({'Neighbourhood':lambda x: list(x)})
```

```python
df['Neighbourhood'].replace('Not assigned',df['Borough'])
```

```python
df_Final = df.groupby(['Postal Code','Borough']).agg({'Neighbourhood':lambda x: list(x)})

df_Final.head()
```
